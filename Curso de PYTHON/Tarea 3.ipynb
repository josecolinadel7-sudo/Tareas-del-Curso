```python
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('ventas_tienda3.xlsx')
```

```python
print(df.columns.tolist())
```

```python
Reporte_Categoria = df.groupby('Categoria', as_index=False)['Ingreso Total'].sum()
```

```python
Reporte_Categoria = Reporte_Categoria.sort_values(by='Ingreso Total', ascending=False)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=Reporte_Categoria, x='Categoria', y='Ingreso Total', palette='viridis')
plt.title('Ingresos Totales por Categoría')
plt.xlabel('Categoría')
plt.ylabel('Ingreso Total')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
Reporte_Categoria.to_excel('Reporte Categorias3.xlsx', index=False)
```
